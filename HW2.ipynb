## Question 1 - List of Tuples

```python
import numpy as np
def lst_of_tup(n, k=3, low=0, high=10):
    out_list = []
    for i in range(n):
        rand_array = np.random.choice(range(low,high), k)
        #rand_array = np.dtype("float32")
        rand_tuple = tuple(rand_array)
        out_list.append(rand_tuple)
    assert isinstance(out_list, list)
    assert isinstance(rand_tuple, tuple)
    return out_list
lst_of_tup(2,4,10,20)
```

## a.

```python
def snippet(sample_list=[(1, 3, 5), (0, 1, 2), (1, 9, 8)], first_index=0, last_index=2):
    '''
    

    Parameters
    ----------
    sample_list : list of tuples
        The default is [(1, 3, 5), (0, 1, 2), (1, 9, 8)].
    first_index : int 
        the first index inside a tuple. The default is 0.
    last_index : int
        the last index inside a tuple. The default is 2.

    Returns
    -------
    res : list of tuples

    '''
    op = []
    for m in range(len(sample_list)):
        li = [sample_list[m]]
        for n in range(len(sample_list)):
            if (sample_list[m][first_index] == sample_list[n][first_index] and
                sample_list[m][last_index] != sample_list[n][last_index]):                    
                li.append(sample_list[n])
        op.append(sorted(li, key=lambda dd: dd[last_index], reverse=True)[first_index])
    res = list(set(op))
    return res

snippet()
```
